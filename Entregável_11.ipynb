```python
class Produto: (define uma classe)
    def __init__(self, nome: str, preco: float): (construtor da classe)
        self.__nome = nome (self é uma referência ao objeto criado)
        self.__preco = preco
        self.__moeda = "BRL"

    # Getters(pega um valor de dentro do objeto)
    def get_nome(self):
        return self.__nome

    def get_preco(self):
        return self.__preco

    def get_moeda(self):
        return self.__moeda

    # Setters(muda um valor dentro do objeto)
    def set_nome(self, nome):
        self.__nome = nome

    def set_preco(self, preco):
        self.__preco = preco

    def set_moeda(self, moeda):
        self.__moeda = moeda

# Método que exibe as informações do produto
    def __str__(self):
        return f"Produto: {self.__nome}, Preço: {self.__preco:.2f} {self.__moeda}"

# Subclasse ProdutoAlimenticio
class ProdutoAlimenticio(Produto):
    def __str__(self):
        return f"[Alimento] {self.get_nome()} - {self.get_preco():.2f} {self.get_moeda()}"


# Subclasse ProdutoEletronico
class ProdutoEletronico(Produto):
    def __str__(self):
        return f"[Eletrônico] {self.get_nome()} - {self.get_preco():.2f} {self.get_moeda()}"


# Exceção personalizada
class MoedaInvalidaError(Exception):
    pass


# Classe de conversão de moedas
class ConversorMoeda:
    def __init__(self):
        self.__usd_brl = 5.05
        self.__eur_brl = 6.14
        self.__eur_usd = 1.22

    def converte_preco_para_usd(self, produto: Produto) -> bool:
        moeda_atual = produto.get_moeda()

        if moeda_atual == "USD":
            return False
        elif moeda_atual == "BRL":
            novo_preco = produto.get_preco() / self.__usd_brl
        elif moeda_atual == "EUR":
            novo_preco = produto.get_preco() / self.__eur_usd
        else:
            raise MoedaInvalidaError("Moeda não reconhecida.")

        produto.set_preco(novo_preco) (muda o preço)
        produto.set_moeda("USD") (muda a moeda)
        return True

    def converte_preco_para_eur(self, produto: Produto) -> bool:
        moeda_atual = produto.get_moeda()

        if moeda_atual == "EUR":
            return False
        elif moeda_atual == "BRL":
            novo_preco = produto.get_preco() / self.__eur_brl
        elif moeda_atual == "USD":
            novo_preco = produto.get_preco() * (1 / self.__eur_usd)
        else:
            raise MoedaInvalidaError("Moeda não reconhecida.")

        produto.set_preco(novo_preco)
        produto.set_moeda("EUR")
        return True

    def converte_preco_para_brl(self, produto: Produto) -> bool:
        moeda_atual = produto.get_moeda()

        if moeda_atual == "BRL":
            return False
        elif moeda_atual == "USD":
            novo_preco = produto.get_preco() * self.__usd_brl
        elif moeda_atual == "EUR":
            novo_preco = produto.get_preco() * self.__eur_brl
        else:
            raise MoedaInvalidaError("Moeda não reconhecida.")

        produto.set_preco(novo_preco)
        produto.set_moeda("BRL")
        return True

def main():
  nome = input(“Informe o nome do produto:”)
  preco = float(“Informe o preço do produto)
  tipo = input(“Informe o tipo do produto(alimentício ou eletrônico”)):

  #objeto das subclasses
  if tipo == “alimentício”:
    produto = ProdutoAlimenticio(nome, preco)
  elif tipo == “eletrônico”:
    produto = ProdutoEletronico(nome, preco)
  else:
        print(“Tipo inválido”)
        produto = Produto(nome, preco)

  print("Produto adicionado:")
  print(produto)

  try:
    if ConversorMoeda.converte_preco_para_usd(produto):
      print("Produto convertido para USD:")
      print(produto)

    else:
      print("O produto  já está em USD")
  except MoedaInvalidError as i:
    print(f"Erro: {i}")


if __name__ == "__main__":
  main()
```
